Here is some simple example code to make a plot.  Four your own plots, you can copy and place different elements of this code and modify as needed.

```python
planet_names=["Mercury", "Venus", "Earth"]    #Fill out with other planets
planet_a=np.array([1,2,3])     #Replace/add the planet semi-major axes here
planet_mass=np.array([100,200,300])   #Replace/add the planet masses here
```

```python
#All plots should have these two lines
#First, we create a "Figure" object
fig=plt.figure(1,figsize=(18,5))  #Numbers in parentheses determine the dimensions of the plot
#Next, we create an "axis" object.  I'll call this "ax1" (that's a number 1, not the letter l)
#If I wanted more than one plot, I might call the next ax2, and the next ax3, etc.
ax1=fig.add_subplot(111)     #11 Means make a grid 1 plot box wide by 1 box tall.  
                             #The final 1 means let ax1 be the first plot box.

#Plot desired x and y points.  'ro' tells python to make the points red (r) circles (o).
ax1.plot(planet_a, planet_mass, 'ro')    

#Now use a loop to label each point with the planet name
for i, (x, y) in enumerate(zip(planet_a, planet_mass)):
    ax1.text(x,y,planet_names[i],fontsize=14)
#Make sure to check that the font is legible on your plot

#When plotting functions, you need to create a number of x and y values and then plot them.
#One convenient way to create a lot of x values is to use numpy's "arange" "linspace" or
#"logspace" functions

#Plot the function y=100x
x=np.arange(0,1000,1)  #Make x a long list of numbers from 0 to 1000, counting by 1
y=100*x                #Let y = 100 times x
ax1.plot(x,y)       #Plot y vs. x.  By default, python will plot a solid line.

ax1.set_yscale('linear')
ax1.set_xscale('linear')
#Change these to 'log' to make a log plot.  Caution!  You do not want to have a 0 in a log plot.
#(Because, what is the log of 0?)

#Label your axes
ax1.set_xlabel('Planet index',fontsize=18)
ax1.set_ylabel('Planet index x 100',fontsize=18)

#Set x and y plot limits
ax1.set_xlim(1,4)   #Set limits for x axis to make the plot look better
ax1.set_ylim(90,400)  #Set limits for y axis to make the plot look better

#Use these lines to increase the font size of the x and y axis labels
for tick in ax1.xaxis.get_major_ticks():   #Change x axis label sizes
    tick.label1.set_fontsize(17)
    tick.label1.set_fontweight('medium')
for tick in ax1.yaxis.get_major_ticks():   #Change y axis label sizes
    tick.label1.set_fontsize(17)
    tick.label1.set_fontweight('medium')
#Check to make sure the font is legible on your final plot
```
